```python
# imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

```python
# reading in
df = pd.read_csv('./CollegeScorecard_Raw_Data/MERGED2015_16_PP.csv')
```

```python
# organize data
column_names = list(df)
ugds_headers = [x for x in column_names if x[:5] == "UGDS_"][:4]
headers = ['INSTNM','STABBR'] + ugds_headers
df_select = df[headers]
df_select = df_select.dropna()
df_state = df_select.groupby(['STABBR']).mean()
df_state
```

```python
# import graphing packages
import plotly
import plotly.plotly as py
import plotly.figure_factory as ff
fips = pd.read_csv('./state_fips.csv')
combined = pd.merge(fips, df_state, left_on='Name', right_on='STABBR')
combined
```

```python
# (ugds - pop) / pop
combined['perc_diff_white'] = (combined['UGDS_WHITE'].astype('float64') - combined['White'].astype('float64'))/combined['White'].astype('float64')
combined['perc_diff_asian'] = (combined['UGDS_ASIAN'].astype('float64') - combined['Asian'].astype('float64'))/combined['Asian'].astype('float64')
combined['perc_diff_black'] = (combined['UGDS_BLACK'].astype('float64') - combined['Black'].astype('float64'))/combined['Black'].astype('float64')
combined['perc_diff_hisp'] = (combined['UGDS_HISP'].astype('float64') - combined['Hispanic'].astype('float64'))/combined['Hispanic'].astype('float64')
combined
```

```python
def plot(race_to_plot, scale, colorbar_title, graph_title):
    data = [ dict(
            type='choropleth',
            colorscale = scale,
            autocolorscale = False,
            locations = combined['Name'],
            z = combined[race_to_plot].astype(float),
            locationmode = 'USA-states',
            colorbar = dict(
                title = colorbar_title
            ),
            marker = dict(
                    line = dict (
                        color = 'rgb(255,255,255)',
                        width = 2
                    )
            )
        ) ]

    layout = dict(
            title = graph_title,
            geo = dict(
                scope='usa',
                projection=dict( type='albers usa' )
            ),
        )

    fig = dict(data=data, layout=layout)
    url = py.plot(fig, filename='d3-cloropleth-map')
```

```python
# plot white
race_to_plot = 'perc_diff_white'
colorbar_title = "% Difference"
graph_title = 'Percent Difference between University and State Population Proportions '
scl = [[0, 'rgb(204,229,255)'],[0.2, 'rgb(153,255,255)'],[0.4, 'rgb(102,178,255)'],\
            [0.6, 'rgb(51,153,255)'],[0.8, 'rgb(0,128,255)'],[1.0, 'rgb(0,102,204)']]

graph_title_white = graph_title + str("(White)")
plot(race_to_plot, scl, colorbar_title, graph_title_white)
```

```python
# plot asian
race_to_plot = 'perc_diff_asian'
graph_title_asian = graph_title + str("(Asian)")
scl = [[0, 'rgb(204,255,229)'],[0.2, 'rgb(153,255,204)'],[0.4, 'rgb(102,255,178)'],\
            [0.6, 'rgb(51,255,153)'],[0.8, 'rgb(0,255,128)'],[1.0, 'rgb(0,204,102)']]

plot(race_to_plot, scl, colorbar_title, graph_title_asian)
```

```python
# plot black
race_to_plot = 'perc_diff_black'
graph_title_black = graph_title + str("(Black)")
scl = [[0.0, 'rgb(229,204,255)'],[0.2, 'rgb(204,153,255)'],[0.4, 'rgb(178,102,255)'],\
            [0.6, 'rgb(153,51,255)'],[0.8, 'rgb(127,0,255)'],[1.0, 'rgb(102,0,204)']]

plot(race_to_plot, scl, colorbar_title, graph_title_black)
```

```python
# plot hispanic
race_to_plot = 'perc_diff_hisp'
graph_title_hisp = graph_title + str("(Hispanic)")
scl = [[0.0, 'rgb(255,229,204)'],[0.2, 'rgb(255,204,153)'],[0.4, 'rgb(255,178,102)'],\
            [0.6, 'rgb(255,153,51)'],[0.8, 'rgb(255,128,0)'],[1.0, 'rgb(204,102,0)']]

plot(race_to_plot, scl, colorbar_title, graph_title_hisp)
```
